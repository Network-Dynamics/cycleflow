## Study of node perturbation in linear flow networks

Let a fixed point in a flow network be

$$\vec{v}\text{ such that }P_j+\sum_j (v_i-v_j)=0$$

If we perturb a single node $n$ like this:
$$P'_j=P_j+\epsilon\delta_{jn}$$

The fixed point will change like this:
$$\vec{v} \mapsto \vec{v} + \delta \vec{v}$$

$$L(G) \vec{\delta v} \approx (0,0,0,q,0,0,0)$$
($q$ at n'th position, $0$ elsewhere)

###Q: How does perturbation scale with distance, in different topologies?

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import operator

from scipy.sparse.linalg import lsqr as solve   #ugly hack, do change
import pylab
pylab.rcParams['figure.figsize'] = (20.0, 8.0)
```

```python
def perturb_vs_dist(G):
    """
    accepts:
        G: a networkx graph
    does:
        perturrbs the centralmost node
    returns:
        the perturbations against shortest path distance from centralmost node in a dictionary format
    """
    L=np.array(nx.laplacian_matrix(G))
       
    centralities=nx.betweenness_centrality(G)
    centralmost_node=max(centralities.items(), key=operator.itemgetter(1))[0]
    whereis_centermost_node=G.nodes().index(centralmost_node)
    
    qvec=np.zeros(G.number_of_nodes())
    qvec[whereis_centermost_node]=1
    
    perturb=solve(L,qvec)[0]
    perturb-=perturb[whereis_centermost_node]
    pertdict={}
    
    distances=nx.shortest_path_length(G,source=centralmost_node)
    
    for idx,other in enumerate(G.nodes()):
        dist=distances[other]
        
        if pertdict.has_key(dist):
            pertdict[dist].append(perturb[idx])
        else:
            pertdict[dist]=[perturb[idx]]
            
    return pertdict


    
```

```python
def plot_mean_perturb_vs_dist(G):
    """
    plots both minnmax and mean perturbs with distance
    returns: dists, mean_perturb
    """
    perturbs=perturb_vs_dist(G)
    
    dists=np.array(perturbs.keys())
    plt.vlines(dists,ymax=[max(perturbs[dist]) for dist in dists], ymin=[min(perturbs[dist]) for dist in dists])
    
    meandata=np.array([np.mean(perturbs[dist]) for dist in dists])
    plt.plot(dists, meandata)
    return dists, meandata
```

###Topology 2: 2d grid graph
-  Expected behaviour: $\delta v \propto 1/d$
-  Regime of validity: $d<<N$


```python
dist_g, mean_g=plot_mean_perturb_vs_dist(nx.grid_2d_graph(50,50))
```

```python
plt.plot(dist_g,mean_g)
plt.plot(dist_g,1/dist_g)
```

###Topology 2: Tree graph
-  Expected behaviour: $\delta v \propto -d$
-  Regime of validity: $d<<N$


```python
dist_t,mean_t=plot_mean_perturb_vs_dist(nx.balanced_tree(2,10))
```

###Topology 3: Ring graph
-  Expected behaviour: $\delta v \approx -d/2$
-  Regime of validity: $d^2/N<<1$

```python
dist_r, mean_r=plot_mean_perturb_vs_dist(nx.cycle_graph(400))
plt.plot(dist_r, -dist_r/2.0)
```
