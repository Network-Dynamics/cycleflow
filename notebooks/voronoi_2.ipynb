```python
import VoronoiGraph as vg
import matplotlib.pyplot as plt
import networkx as nx
%matplotlib inline
import pylab
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
gr=vg.VoronoiPlanarGraph(15)
```

```python
G=gr.graph
H=gr.dual_graph
```

```python
nx.draw(G, pos={n:G.node[n]['pos'] for n in G.nodes()})
nx.draw(H, pos={n:H.node[n]['pos'] for n in H.nodes()}, node_color='g', edge_color='b')
```
