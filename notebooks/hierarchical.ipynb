## The question:

The spread of **fluctuation in flows** (to be defined precisely later) decreases with **distance** from the source of perturbation.
How does the decay scale: exponential or power law?

### First we look at cycle flow changes vs. distance

```python
import matplotlib.pyplot as plt
import networkx as nx
%matplotlib inline
import matplotlib as mpl
from __future__ import division
from math import sqrt
```

```python
#%%writefile HierarchicalGraph.py
class HierarchicalGraph:
    def __init__(self, maxdepth, depth, weights="mult", compute_dual=False):
        self.size=2**(maxdepth-1)+1 #The lattice size
        self.graph=self._hierarchical(maxdepth, depth, weights)        
        
        if compute_dual:
            self.dual_graph=self.get_dual_graph()
        
    def get_dual_graph(self):
        """
        Creates the *loopy* dual graph
        """
        H=nx.MultiGraph()
        
        #This takes care of everything except the self loops and the right/top edges
        for x in range(self.size-2):
            for y in range(self.size-2):
                newnode=(x+0.5, y+0.5)
                H.add_edge((x+0.5,y+0.5),(x+1.5,y+0.5),weight=self.graph[(x+1,y)][(x+1,y+1)]['weight'])
                H.add_edge((x+0.5,y+0.5),(x+0.5,y+1.5),weight=self.graph[(x,y+1)][(x+1,y+1)]['weight'])
                
        for x in range(0, self.size-1):
            y=x
            
            ##Now adding self loops. Making the weight  /2 because then we will have the correct degree
            #left self loops
            H.add_edge((0.5,y+0.5), (0.5,y+0.5), weight=self.graph[(0,y)][(0,y+1)]['weight']/2)
            
            #right self loops
            H.add_edge((self.size-1.5,y+0.5), (self.size-1.5,y+0.5), weight=self.graph[(self.size-1,y)][(self.size-1,y+1)]['weight']/2)
            
            #bottom self loops
            H.add_edge((x+0.5,0.5), (x+0.5,0.5), weight=self.graph[(x,0)][(x+1,0)]['weight']/2)
            
            #top self loops
            H.add_edge((x+0.5,self.size-1.5), (x+0.5,self.size-1.5), weight=self.graph[(x,self.size-1)][(x+1,self.size-1)]['weight']/2)
            
        for x in range(self.size-2):
            y=x
            #top edge
            H.add_edge((x+0.5,self.size-1.5), (x+1.5,self.size-1.5), weight=self.graph[(x+1,self.size-1)][(x+1,self.size-2)]['weight'])
            
            #right edge
            H.add_edge((self.size-1.5,y+0.5), (self.size-1.5,y+1.5), weight=self.graph[(self.size-2,y+1)][(self.size-1,y+1)]['weight'])

        assert(H.number_of_edges()==self.graph.number_of_edges())
        return H        
            
    def _hierarchical(self,maxdepth, depth, weights="mult"):
        """ 
        Creates a hierarchical network with depth number of hierarchy level
        """ 
        assert(depth<maxdepth)
        size=2**(maxdepth-1)+1
         
        G=nx.grid_2d_graph(size, size)
         
        stepsize=2**(maxdepth-depth-1)
         
        baseweight=1
        weight=baseweight
        d=1 
         
        for u,v in G.edges():
            G[u][v]['weight']=weight
            G[v][u]['weight']=weight
             
         
        while d<=depth:
            #Build up the weights, thinnest first
            for coord in range(0,size, stepsize):
                for other in range(size-1):
                    G[(coord, other)][(coord, other+1)]['weight']=weight
                    G[(coord, other+1)][(coord, other)]['weight']=weight
                     
                    G[(other, coord)][(other+1, coord)]['weight']=weight
                    G[(other+1, coord)][(other, coord)]['weight']=weight
     
     
            if weights=='mult':
                weight*=2
            else:
                weight+=baseweight
                 
            stepsize*=2
            d+=1
        return G
```

## 1. Create The graph

```python
HG=HierarchicalGraph(8,6, compute_dual=True, weights='add')
G=HG.graph
H=HG.dual_graph
```

## 2. Now compute the cycleflow changes

```python
from scipy.sparse.linalg import spsolve
from scipy.sparse import diags
def cycleflow_changes(HG, perturb=None):
    dg=HG.get_dual_graph()
    
    degrees=nx.degree(dg, weight='weight')
    loopy_lapl= diags([[degrees[node] for node in dg.nodes()]], [0], format='csr') \
    - nx.to_scipy_sparse_matrix(dg, weight='weight', format='csr')
    
    size=HG.size
    mid=size//2

    if perturb is None:
        #Perturb at main vein, as a result, two cycles get affected
        perturb=((mid-0.5, mid+0.5), (mid+0.5, mid+0.5))
    
    srcvec=np.zeros(dg.number_of_nodes())
    
    allnodes_idx={node:idx for idx, node in enumerate(dg.nodes())}
    
    srcvec[allnodes_idx[perturb[0]]]=1
    srcvec[allnodes_idx[perturb[1]]]=-1
    
    df_c=solve(loopy_lapl.toarray(), srcvec)
    return {node:df_c[allnodes_idx[node]] for node in dg.nodes()}
   
```

```python
dcfs_dict=cycleflow_changes(HG)
```

## 3. Scaling analysis:

### Way to gather all the y points for *each* x point:

```python
def aggregate(xdata, ydata):
        """
        arguments:
                xdata: python iterable
                ydata: python iterable
        xdata and ydata must have equal length

        returns: {xpt1: [yval1, yval2,...], xpt2: [yval1, yval2,...]}
                xptn: unique elements of xdata, in increasing order
                yvaln: yvalue for that xpt
        """
        xdata=np.array(xdata)
        ydata=np.array(ydata)
                
        xpoints=sorted(set(xdata))
        
        data=[]

        for x in xpoints:
                yvals_for_x=ydata[xdata==x]
                data.append(yvals_for_x)
                
    
        return np.array(list(xpoints)), data
```

### 3.a Scaling against euclidean distance

```python
dcf_arr=[abs(dcfs_dict[node]) for node in HG.dual_graph.nodes()]
eucl_dists_arr=[sqrt((x-HG.size//2)**2+(y-HG.size//2-0.5)**2) for (x,y) in HG.dual_graph.nodes()]

xcoords, data=aggregate(eucl_dists_arr, dcf_arr)
```

```python
from matplotlib import cbook as cbook
res=plt.violinplot(data, positions=xcoords, showmeans=True)


means=np.array([elem[0,1] for elem in res['cmeans'].get_segments()])
maxes=np.array([elem[0,1] for elem in res['cmaxes'].get_segments()])
mins=np.array([elem[0,1] for elem in res['cmins'].get_segments()])
```

```python
plt.plot(xcoords, means, 'g-')
plt.plot(xcoords, maxes, 'r-')
plt.plot(xcoords, mins, 'b-')
```

### 3.b Scaling against cycle distances:

```python
def cycle_distance((x,y), src=((0,0),(0,0))):
    (x_a, y_a),(x_b,y_b)=src
    
    return (abs(x-x_a)+abs(y-y_a)+abs(x-x_b)+abs(y-y_b))/2
    
```

```python
size=HG.size

dcf_arr=[abs(dcfs_dict[node]) for node in HG.dual_graph.nodes()]

cycl_dists_arr=[cycle_distance(node, src=((size//2-0.5, size//2+0.5),(size//2+0.5, size//2+0.5))) for node in HG.dual_graph.nodes()]

xcoords, data=aggregate(cycl_dists_arr, dcf_arr)
```

```python
res=plt.violinplot(data, positions=xcoords, showmeans=True)


means=np.array([elem[0,1] for elem in res['cmeans'].get_segments()])
maxes=np.array([elem[0,1] for elem in res['cmaxes'].get_segments()])
mins=np.array([elem[0,1] for elem in res['cmins'].get_segments()])
```

```python
plt.plot(xcoords, means, 'g-')
plt.plot(xcoords, maxes, 'r-')
plt.plot(xcoords, mins, 'b-')
```

#### Is it exponential?

```python
plt.plot(xcoords, np.log(means), 'g-')
plt.plot(xcoords, np.log(maxes), 'r-')
plt.plot(xcoords, np.log(mins), 'b-')
```
