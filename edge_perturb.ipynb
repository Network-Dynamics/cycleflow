#Studying the effect of perturbing an edge in a Kuramoto oscillator network


Let's say we have a dynamical system:
$$\dot{\vec{\theta}}=\vec{P}+M\sin{\left(M^T\vec{\theta}\right)}$$

And a fixed point:
$$\vec{P}+M\sin{\left(M^T\vec{\theta^*}\right)}=0$$

If we change the weight of only one edge like this:
$$w_{i,j}\mapsto w_{i,j}+\epsilon$$

Then the the fixed point will look like this:
$$\vec{\theta'}=\vec{\theta^*}+\delta\vec{\theta}$$
$$L(D(G))\vec{\delta\theta}=(0,0,\cdots,q,-q,0,0,\cdots)$$
$$q=\epsilon\sin{(\theta^*_i-\theta^*_j)}$$

```python
import kuramoto
```

```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

## Let's setup our powergrid:

```python
nnodes=100
P_coeff=0.1 #power production +=1
p_rew=0.3
```

```python
G=nx.random_graphs.erdos_renyi_graph(nnodes, p_rew)
```

```python
P=np.ones(nnodes)*P_coeff
P[np.random.choice(range(nnodes), size=nnodes/2, replace=False)]*=-1 #half of them are consumers
```

```python
nx.draw(G)
```

```python
M=np.array(nx.incidence_matrix(G, oriented=1))
```

##Let's simulate

```python
from scipy.integrate import odeint
reload(kuramoto)
```

```python
M.shape
```

```python
fp, details=kuramoto.try_find_fps(10,M,M,P)
```

####Let's store the Laplacian of meta graph

```python
res_cap_edges=np.cos(np.dot(M.T, fp))
```

```python
L=np.array(nx.laplacian_matrix(G))
for m in range(nnodes):
    for n in range(m+1,nnodes):
        res_cap=res_cap_edges[np.cos(fp[m]-fp[n])]
        L[n,m]=res_cap
        L[m,n]=res_cap
```

###Let's perturb

```python
import operator
epsilon=0.1
```

Get the centralmost edge

```python
centralities=nx.edge_betweenness_centrality(G)
(u,v)=max(centralities.iteritems(), key=operator.itemgetter(1))[0]
```

```python
u,v
```

Change the inclidence matrices

```python
M_w=M
```

```python
M_w[:,G.edges().index((u,v))]*=(1+epsilon)
```

Q vector

```python
charge=epsilon*(np.sin(fp[u]-fp[v]))
```

#####New solution

```python
fpnew, details=kuramoto.try_find_fps(10,M,M,P)
```

```python
differences=(fpnew-fpnew[0])-(fp-fp[0])
```

### Check against prediction:

```python
qbar=np.dot(L, differences)
```

```python
qbar=qbar-np.average(qbar)
```
